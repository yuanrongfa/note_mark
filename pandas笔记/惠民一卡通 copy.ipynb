```python
import pandas as pd
import numpy as np
import os

# 存放数据的路径
path = r"E:\惠民一卡通\农信社6.29\都安县"

# 修改文件后缀为xlsx
a = os.listdir(path)
for i in a:
    if i[-4:] == ".xls":
        os.rename(path + f"\\{i}", path + f"\\{i}x")
```

# 开始处理

```python
per_col = ['序号', '*补贴名称', '*享受补贴人姓名', '*享受补贴人身份证号码', '*县(区)', '*镇(乡)', '*村(社区)', '*组', '*银行卡号', '*所属银行', '*持卡人', '*持卡人与享受补贴人的关系']

data = pd.DataFrame()
diff_list = []
a = os.listdir(path)
for i in a:
    print(i)
    if i == "格式化汇总.xlsx" or i == "相差.xlsx" or i[-4:] == ".txt":
        continue
    num_l = []
    num_l.append(i)
    d = pd.read_excel(path + f"\\{i}",
    dtype={"序号": np.str_,"*享受补贴人身份证号码": np.str_, "*银行卡号": np.str_})
    num_l.append(len(d))

    # 预处理
    d.dropna(inplace=True, thresh=12)
    d.dropna(axis=1, inplace=True, how="all")
    d = d.iloc[:, :12]
    try:
        d.columns = per_col
    except ValueError as ve:
        print(i + "------------------------------------------------ 缺少列！")
        continue

    # d.rename(columns={list(d)[0]: "序号"}, inplace=True)
    # d.columns = per_col

    d["序号"] = d["序号"].astype(str)
    d["序号"] = d["序号"].str.replace(' ', '')
    if d["序号"].iloc[0] == "1":
        pass
    elif d["序号"].iloc[0] == "序号":
        d = d.iloc[1:]

    num_l.append(len(d))
    num_l.append(num_l[1] - num_l[-1])

    try:
        if d["序号"].iloc[0] != "1":
            num_l.append(1)
        else:
            num_l.append(0)
    except IndexError as ie:
        print(i + "------------------------------------------------ 数据为空！")
        num_l.append(1)
        continue
    diff_list.append(num_l)
    data = pd.concat([data, d])

# 添加占位符    
data["占位符"] = True
# 修复序号
data["序号"] = range(1,len(data) + 1)
print("--------------------")
print(len(a))
print(f"data shape {data.shape}")
```

##### 输出相差情况

```python
diff_data = pd.DataFrame(diff_list)
diff_data.columns = ["文件名", "原始数", "删除缺失数", "相差数", "序号不为1"]
diff_data.sort_values(["序号不为1", "相差数"], ascending=False, inplace=True)
diff_data.to_excel(path + "\\相差.xlsx", index=False)
```
